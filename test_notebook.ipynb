```python
# Reserve an order ID sequence
baseId = ib.client.getReqId()

parent = MarketOrder('BUY', 1)
parent.orderId = baseId
parent.transmit = False

sl = StopOrder(
    action='SELL',
    totalQuantity=1,
    stopPrice=24500,
    parentId=baseId,
    orderId=baseId + 1,
    tif='GTC',
    transmit=False
)

tp = LimitOrder(
    action='SELL',
    totalQuantity=1,
    lmtPrice=30000,
    parentId=baseId,
    orderId=baseId + 2,
    tif='GTC',
    transmit=True
)

for o in [parent, sl, tp]:
    ib.placeOrder(contract, o)
```
